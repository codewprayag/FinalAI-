# Question 2: Build a Model from Scratch

```python
import numpy as np
X = df[['GDP']].values

y = df['Life expectancy'].values.reshape(-1, 1)

import numpy as np

# Shuffle the dataset
np.random.seed(42)  # Ensures reproducibility
indices = np.random.permutation(len(X))  # Shuffle indices

# Define train-test split (80% train, 20% test)
train_size = int(0.8 * len(X))
train_idx, test_idx = indices[:train_size], indices[train_size:]

# Split the data
X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

print("Data split complete. Train size:", len(X_train), "Test size:", len(X_test))
```

```python
# Ensure X_train and X_test are NumPy arrays
X_train_np = np.array(X_train)
X_test_np = np.array(X_test)

# Add bias term (column of ones) for the intercept
X_train_bias = np.c_[np.ones((X_train_np.shape[0], 1)), X_train_np]
X_test_bias = np.c_[np.ones((X_test_np.shape[0], 1)), X_test_np]

# Compute optimal weights using the Normal Equation
theta = np.linalg.inv(X_train_bias.T @ X_train_bias) @ X_train_bias.T @ y_train

# Make predictions
y_train_pred_manual = X_train_bias @ theta
y_test_pred_manual = X_test_bias @ theta

# Compute Mean Squared Error (MSE)
mse_train_manual = np.mean((y_train - y_train_pred_manual) ** 2)
mse_test_manual = np.mean((y_test - y_test_pred_manual) ** 2)

# Compute R-squared (R²)
ss_total_train = np.sum((y_train - np.mean(y_train)) ** 2)
ss_residual_train = np.sum((y_train - y_train_pred_manual) ** 2)
r2_train_manual = 1 - (ss_residual_train / ss_total_train)

ss_total_test = np.sum((y_test - np.mean(y_test)) ** 2)
ss_residual_test = np.sum((y_test - y_test_pred_manual) ** 2)
r2_test_manual = 1 - (ss_residual_test / ss_total_test)

print("MSE train:",mse_train_manual);
print("R-squared train:",r2_train_manual);
print("MSE test:",mse_test_manual);
print( "R-squared",r2_test_manual);
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Function to generate polynomial features
def generate_polynomial_features(X, degree):
    """
    Expands the feature matrix X to include polynomial terms up to the given degree.
    """
    X_poly = X.copy()
    for d in range(2, degree + 1):  # Start from x^2 up to x^degree
        X_poly = np.c_[X_poly, X ** d]  # Concatenate new features
    return X_poly

# Choose polynomial degree (e.g., 2 for quadratic regression)
degree = 2

# Scale the dataset before generating polynomial features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Generate polynomial features for train and test sets
X_train_poly = generate_polynomial_features(X_train_scaled, degree)
X_test_poly = generate_polynomial_features(X_test_scaled, degree)

# Add bias term (column of ones) for the intercept
X_train_poly_bias = np.c_[np.ones((X_train_poly.shape[0], 1)), X_train_poly]
X_test_poly_bias = np.c_[np.ones((X_test_poly.shape[0], 1)), X_test_poly]

# Compute optimal weights using the Normal Equation
theta_poly = np.linalg.inv(X_train_poly_bias.T @ X_train_poly_bias) @ X_train_poly_bias.T @ y_train

# Make predictions
y_train_pred_poly = X_train_poly_bias @ theta_poly
y_test_pred_poly = X_test_poly_bias @ theta_poly

# Compute Mean Squared Error (MSE)
mse_train_poly = np.mean((y_train - y_train_pred_poly) ** 2)
mse_test_poly = np.mean((y_test - y_test_pred_poly) ** 2)

# Compute R-squared (R²)
ss_total_train_poly = np.sum((y_train - np.mean(y_train)) ** 2)
ss_residual_train_poly = np.sum((y_train - y_train_pred_poly) ** 2)
r2_train_poly = 1 - (ss_residual_train_poly / ss_total_train_poly)

ss_total_test_poly = np.sum((y_test - np.mean(y_test)) ** 2)
ss_residual_test_poly = np.sum((y_test - y_test_pred_poly) ** 2)
r2_test_poly = 1 - (ss_residual_test_poly / ss_total_test_poly)

# Return both models' results
print("Linear Regression:")
print("MSE train:", mse_train_poly)
print("R-squared train:", r2_train_poly)
print("MSE test:", mse_test_poly)
print("R-squared test:", r2_test_poly)
```

# 4. Hyper-parameter Optimization with Cross-Validation

1. Identify the hyperparameters of the models used in Task 3 - Build a Primary Model (for both

In our regression models, the key hyperparameters that influence performance include the polynomial degree (d) for Polynomial Regression and potential regularization parameters if extensions like Ridge or Lasso Regression were used. For Linear Regression, no significant hyperparameters need tuning since we used the Normal Equation for optimization. However, in Polynomial Regression, the choice of degree (d) is critical, as a higher degree increases model complexity and may improve fit but risks overfitting. To determine the best hyperparameters, cross-validation can be applied to systematically evaluate different values of d and select the optimal degree that balances bias and variance.

2. Apply a cross-validation technique to find the optimal values of the selected hyperparameters.

• Hint: You can use techniques like GridSearchCV or RandomizedSearchCV.

```python
import numpy as np

# Function to generate polynomial features
def generate_polynomial_features(X, degree):
    X_poly = X.copy()
    for d in range(2, degree + 1):
        X_poly = np.c_[X_poly, X ** d]  # Add polynomial terms
    return X_poly

# Function to compute Mean Squared Error (MSE)
def compute_mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Function to perform k-Fold Cross-Validation
def cross_validation(X, y, degrees, k=5):
    fold_size = len(X) // k
    mse_scores = {}

    for degree in degrees:
        mse_list = []

        for i in range(k):
            # Split data into training and validation sets
            start, end = i * fold_size, (i + 1) * fold_size
            X_val, y_val = X[start:end], y[start:end]
            X_train = np.vstack((X[:start], X[end:]))
            y_train = np.vstack((y[:start], y[end:]))

            # Generate polynomial features
            X_train_poly = generate_polynomial_features(X_train, degree)
            X_val_poly = generate_polynomial_features(X_val, degree)

            # Add bias term
            X_train_poly_bias = np.c_[np.ones((X_train_poly.shape[0], 1)), X_train_poly]
            X_val_poly_bias = np.c_[np.ones((X_val_poly.shape[0], 1)), X_val_poly]

            # Train the model using the Normal Equation
            theta = np.linalg.inv(X_train_poly_bias.T @ X_train_poly_bias) @ X_train_poly_bias.T @ y_train

            # Make predictions
            y_val_pred = X_val_poly_bias @ theta

            # Compute MSE
            mse_list.append(compute_mse(y_val, y_val_pred))

        # Store the average MSE for this degree
        mse_scores[degree] = np.mean(mse_list)

    return mse_scores

# Define degrees to test
degrees_to_test = [1, 2, 3, 4, 5]

# Perform cross-validation
mse_results = cross_validation(X_train_scaled, y_train, degrees_to_test)

# Find the best degree (lowest MSE)
best_degree = min(mse_results, key=mse_results.get)

# Print results
print("Cross-Validation Results (MSE per Degree):", mse_results)
print("Best Polynomial Degree:", best_degree)
```
